```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import io

dfs = []
for filename in uploaded.keys():
  if filename.endswith('.csv'):
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

combined_df
```

```python
combined_df.head()
```

```python
print("Missing values per City:")
print(combined_df.isnull().sum())
```

```python
print("Missing values per Station:")
print(combined_df.groupby('station').apply(lambda x: x.isnull().sum()))
```

```python
print("Total missing values per Station:")
print(combined_df.groupby('station').apply(lambda x: x.isnull().sum().sum()))
```

```python
for column in combined_df.columns:
  if column != 'wd':  # Exclude the 'wd' column
    if combined_df[column].isnull().any():
      combined_df[column].fillna(combined_df[column].mean(), inplace=True)
```

```python
print("Missing values per Station:")
print(combined_df.groupby('station').apply(lambda x: x.isnull().sum()))
```

```python
print("Total missing values per Station:")
print(combined_df.groupby('station').apply(lambda x: x.isnull().sum().sum()))
```

```python
if 'wd' in combined_df.columns:
  combined_df = combined_df.drop('wd', axis=1)
```

```python
combined_df.head(10)
```

```python
# prompt: combine year month and day to make a date column

# Assuming you have columns named 'year', 'month', and 'day'
combined_df['date'] = pd.to_datetime(combined_df[['year', 'month', 'day']])
```

```python
combined_df.head(10)
```

```python
# prompt: combine rows based on date and station column and removing hour column

# Group by 'date' and 'station', then aggregate other columns (e.g., by mean)
combined_df = combined_df.groupby(['date', 'station']).agg({
    'PM2.5': 'mean',
    'PM10': 'mean',
    'SO2': 'mean',
    'NO2': 'mean',
    'CO': 'mean',
    'O3': 'mean',
    'TEMP': 'mean',
    'PRES': 'mean',
    'DEWP': 'mean',
    'RAIN': 'mean',
    'WSPM': 'mean',
}).reset_index()

combined_df
```
